# Basic Stats: Box Plot Generator #
## Overview ##
This notebook will allow you to quickly generate boxplots for a dataset loaded in CSV format.
The CSV should consist of columns for each sample set, and a row for each sample. For an example please take a look at the Monthly.csv included in this directory and loaded below.

## Note ##
If you are on a mac and exporting a CSV from excel, please make sure to use CSV rather than the UTF8 CSV. Excel embeds the windows newline character rather than the unix one, which will make pandas barf with the code below. 

```python
# First, import the libraries that will be useful for this: 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
datafile = 'Monthly.csv'
inputdata = pd.read_csv(datafile, sep=',',low_memory=False)
v1 = pd.DataFrame(data=inputdata)
```

## Examining the Data ##
This section is optional- it allows you to take a look at the data, sort it and edit it as needed.

```python
import qgrid
qgrid.nbinstall(overwrite=True)
pd.set_option('display.max_rows', 8)
```

```python
qgrid.show_grid(inputdata)
```

## The plot itself ##
Now that you've loaded the data and examined it, generate the boxplots and see how they look

```python
v1.plot.box(figsize=(8, 6))
# Below settings may provide useful formatting
# plt.xticks(labels, rotation=90)
plt.title('Monthly Percentages of Total Annual Margin \n Actuals from 2006-2015', fontsize=14)
plt.ylabel('Percentage')
plt.xlabel('Month')
plt.grid(True)
```
